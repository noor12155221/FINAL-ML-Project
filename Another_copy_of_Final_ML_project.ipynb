```python
# =============================
# 1️⃣ استيراد المكتبات الأساسية
# =============================
import os

# =============================
# 2️⃣ إعداد اسم الفولدر ومسار الريبو
# =============================
repo_folder = "Final_ML_project"  # اسم الفولدر على Colab
git_repo = "https://github.com/noor12155221/FINAL-ML-Project.git"  # رابط الريبو على GitHub

# =============================
# 3️⃣ حذف الفولدر إذا موجود لتجنب أي مشاكل
# =============================
if os.path.exists(repo_folder):
    print(f"Found existing folder '{repo_folder}', removing it...")
    !rm -rf {repo_folder}

# =============================
# 4️⃣ عمل clone للريبو داخل الفولدر
# =============================
print(f"Cloning repository into '{repo_folder}'...")
!git clone {git_repo} {repo_folder}

# =============================
# 5️⃣ الانتقال إلى الفولدر اللي فيه الريبو
# =============================
%cd {repo_folder}
print(f"Current working directory: {os.getcwd()}")

# =============================
# 6️⃣ تأكيد الملفات
# =============================
print("Files in repository:")
!ls
```
